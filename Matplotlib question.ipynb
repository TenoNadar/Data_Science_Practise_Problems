# Basic plots with Matplotlib

## Matplotlib

Q. Perform a line plot, where x axis value is(1,2,3,5) and y axis value is(2,4,1,3) **(2 mark)**

```python
## START CODE HERE : 
import matplotlib.pyplot as plt
x=[1,2,3,5]
y=[2,4,1,3]
_=plt.plot(x,y)
plt.show()
## END CODE
```

## Scatter plot

Q. Perform a scatter plot, where x axis value is(10,35,15,25,50) and y axis vlaue is(1.1,3.5,1.5,2.5,5.0) **(2 mark)**

```python
## START CODE HERE : 
x=[10,35,15,25,50]
y=[1.1,3.5,1.5,2.5,5.0]
_=plt.plot(x,y,marker='.',linestyle="none")
plt.show()
## END CODE
```

## Histogram

> You can use **help(plt.hist)** to know more about the hist function 

Q. Plot a histogram, where values are (20, 45, 75, 90, 100, 105, 130, 145) and keepin the bins as 6. **(2 marks)**

```python
## START CODE HERE : 
x=[20, 45, 75, 90, 100, 105, 130, 145]
_=plt.hist(x,bins=6)
plt.show()
## END CODE
```

Solve the following Plot line questions using matplotlib where:

Q1. x axis value is(10, 20, 30, 40) and y axis value is(20, 40, 10, 30)

Q2. Add x , y axis labels stating "x-axis" and "y-axis" respectively.**(1 mark)**

Q3. Add title as "Matplotlib". **(1 mark)**

Q4. Add double ticks in y axis with values (0, 2, 4, 6) **(1 mark)**

Q5. Add another value x = (10, 20, 30, 40) and y = (40, 10, 30, 20) and show it. **(1 mark)**

```python
## 1. START CODE HERE : 
x=[10, 20, 30, 40]
y=[20, 40, 10, 30]
## END CODE

## 2. START CODE HERE : 
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel('Y-Axis')
## END CODE

## 3. START CODE HERE : 
plt.title("Matplotlib")
## END CODE## 

## 4. START CODE HERE : 
plt.yticks([20,40,10,30],[0, 2, 4, 6])
## END CODE

## 5. START CODE HERE : 
x1=x +[10, 20, 40, 30]
y1=y+[40, 10, 30, 20]
plt.plot(x1,y1)
plt.show()
## END CODE
```
