# Basic plots with Matplotlib

## Matplotlib

Q. Perform a line plot, where x axis value is(1,2,3,5) and y axis value is(2,4,1,3) **(2 mark)**

```python
## START CODE HERE : 

import matplotlib.pyplot as plt


X=[1,2,3,5]
Y=[2,4,1,3]

plt.plot(X,Y)
plt.show()

## END CODE
```

## Scatter plot

Q. Perform a scatter plot, where x axis value is(10,35,15,25,50) and y axis vlaue is(1.1,3.5,1.5,2.5,5.0) **(2 mark)**

```python
## START CODE HERE : 

import matplotlib.pyplot as plt


X=[10,35,15,25,50]
Y=[1.1,3.5,1.5,2.5,5.0]

plt.scatter(X,Y)
plt.show()


## END CODE
```

## Histogram

> You can use **help(plt.hist)** to know more about the hist function 

Q. Plot a histogram, where values are (20, 45, 75, 90, 100, 105, 130, 145) and keepin the bins as 6. **(2 marks)**

```python
## START CODE HERE : 

import matplotlib.pyplot as plt


X=[20,45,75,90,100,105,130,145]

plt.hist(X,bins=6)
plt.show()

## END CODE
```

```python
## 2. START CODE HERE : 

import matplotlib.pyplot as plt


X=[10,20,30,40]
Y=[20,40,10,30]

plt.plot(X,Y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

## END CODE
```

```python
## 3. START CODE HERE : 

import matplotlib.pyplot as plt


X=[10,20,30,40]
Y=[20,40,10,30]

plt.plot(X,Y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Matplotlib")
plt.show()

## END CODE## 
```

```python
## 4. START CODE HERE : 

import matplotlib.pyplot as plt


X=[10,20,30,40]
Y=[20,40,10,30]

plt.plot(X,Y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Matplotlib")
plt.yticks([20,40,10,30],[0,2,4,6])
plt.show()

## END CODE
```

```python
## 5. START CODE HERE : 

import matplotlib.pyplot as plt


X=[10,20,30,40]
Y=[20,40,10,30]
X1=X+[10,20,30,40]
Y1=Y+[40,10,30,20]

plt.plot(X1,Y1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Matplotlib")
plt.yticks([20,40,10,30],[0,2,4,6])
plt.show()


## END CODE
```
